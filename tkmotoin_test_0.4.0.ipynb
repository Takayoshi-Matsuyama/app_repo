# 正弦波コントローラ

```python
#【正弦波コントローラ】モーションフロー作成
import pprint
import tkmotion as tkm
flow = tkm.MotionFlow()
flow.load_discrete_time()
flow.load_motion_profile()  # loads default profile

# 正弦波コントローラ
sinusoidal_ctrl_index = 4
flow.load_controller(filepath="tkmotion/ctrl/default_controller_config.json", ctrl_index=sinusoidal_ctrl_index)
pprint.pprint(type(flow.controller))
pprint.pprint(flow.controller.get_config())

# 質量・ダンパ・ばねモデル
plant_index=1
phyobj_index=0
flow.load_plant(filepath="tkmotion/plant/default_plant_config.json", plant_index=plant_index, phyobj_index=phyobj_index)
```

```python
#【正弦波コントローラ】MotionFlow実行
mtnflow_df = flow.execute()
mtnflow_df
```

```python
# 【正弦波コントローラ】指令推力の確認
import plotly.express as px
fig = px.line(mtnflow_df, x="time_s", y=["force_N"], title="Force")
fig.update_layout(
    xaxis=dict(title="time [s]"),
    yaxis=dict(title="force [N]"))
fig.show()
```

```python
#【正弦波コントローラ】MotionFlow実行結果の可視化
import plotly.express as px
fig = px.line(
    mtnflow_df,
    x="time_s",
    y=["obj_velocity_m_s", "obj_position_m"],
    title="Motion Profile Execution Result")
fig.update_layout(
    yaxis2=dict(
        title="Position (m)",  # 第2Y軸のタイトル
        overlaying="y",        # 第1Y軸の上に重ねる
        side="right",          # 右側に表示
        showgrid=False         # オプション: グリッド線を非表示
    ),
)
for trace in fig.data:
    # 第2Y軸のトレース更新 / Update 2nd Y-axis trace
    if trace.name == "obj_position_m":
        trace.update(yaxis="y2")
        break
fig.update_yaxes(title_text="Velocity (m/s)", row=1, col=1)
fig.update_xaxes(title_text="Time (s)", row=1, col=1)
fig.show()
```

# FFT

スイープ周波数
1. スイープ開始周波数 f_start = 0.1 [Hz]
2. スイープ終了周波数 f_end = 10 [Hz]
3. 周波数分解能 df_req = 0.1 [Hz]

時系列データ要件
1. 周波数分解能 --> 最低限必要な時間<br>
T_req = 1 / df_req = 1 / 0.1 = 10 [s]  # 最小（最遅）周波数の波 1個分
2. 最大分析周波数 --> 最低限必要なサンプリング周波数 f_req、時間刻み dt_req<br>
f_req = f_end * 2.56 = 10 * 2.56 = 25.6 [Hz]  # 最大（最速）周波数の2.56倍(※1)<br>
 (※1) 2.56: エイリアシングの影響を低減するために、実際に使われる倍率<br>
dt_req = 1 / f_req = 1 / 25.6 = 0.039625 [s]
3. 最低限必要なデータ点数<br>
N_req = T_req * f_req = 10 * 25.6 = 256 [個]
4. FFT要件 2のn乗<br>
256 = 2^8 (そのまま)

Note: 周波数 f [Hz] = 単位時間当たりのデータ個数 n [1/s]

```python
#【FFT】モーションフロー作成
import pprint
import tkmotion as tkm
flow = tkm.MotionFlow()
flow.load_discrete_time()
flow.discrete_time.dt = 1 / 25.6
flow.discrete_time.duration = 10.0
print(f"{flow.discrete_time.dt=}s")
print(f"{flow.discrete_time.duration=}s")
flow.load_motion_profile()

# サインスイープコントローラ
sinsweep_ctrl_index = 5
flow.load_controller(filepath="tkmotion/ctrl/default_controller_config.json", ctrl_index=sinsweep_ctrl_index)
flow.controller.f_start = 0.1  # 開始周波数 [Hz]
flow.controller.f_end = 10.0   # 終了周波数 [Hz]
flow.controller.T = flow.discrete_time.duration
print(f"{flow.controller.f_start=}Hz")
print(f"{flow.controller.f_end=}Hz")
print(f"{flow.controller.T=}s")

# 質量・ダンパ・ばねモデル
plant_index=1
phyobj_index=0
flow.load_plant(filepath="tkmotion/plant/default_plant_config.json", plant_index=plant_index, phyobj_index=phyobj_index)
```

```python
#【FFT】MotionFlow実行
mtnflow_df = flow.execute()
mtnflow_df
```

```python
# 【FFT】指令推力の確認
import plotly.express as px
fig = px.line(mtnflow_df, x="time_s", y=["force_N"], title="Force")
fig.update_layout(
    xaxis=dict(title="time [s]"),
    yaxis=dict(title="force [N]"))
fig.show()
```

```python
#【FFT】MotionFlow実行結果の可視化
import plotly.express as px
fig = px.line(
    mtnflow_df,
    x="time_s",
    y=["obj_velocity_m_s", "obj_position_m"],
    title="Motion Profile Execution Result")
fig.update_layout(
    yaxis2=dict(
        title="Position (m)",  # 第2Y軸のタイトル
        overlaying="y",        # 第1Y軸の上に重ねる
        side="right",          # 右側に表示
        showgrid=False         # オプション: グリッド線を非表示
    ),
)
for trace in fig.data:
    # 第2Y軸のトレース更新 / Update 2nd Y-axis trace
    if trace.name == "obj_position_m":
        trace.update(yaxis="y2")
        break
fig.update_yaxes(title_text="Velocity (m/s)", row=1, col=1)
fig.update_xaxes(title_text="Time (s)", row=1, col=1)
fig.show()
```

```python
#【FFT】FFT分析
from scipy.fft import rfft, rfftfreq

# Number of sample points
N = len(mtnflow_df["time_s"])
print(f"{N=}")

# sample spacing
d = flow.discrete_time.dt  # [s]
print(f"{d=}")

# spectrum : complex ndarray (a: array_like)
yf_r = rfft(mtnflow_df["obj_position_m"])
print(f"{type(yf_r)=}")
print(f"{type(yf_r[0])=}")
print(f"{yf_r.shape=}")

# 振幅正規化（片側スペクトル）
yf_r_norm = np.abs(yf_r) / (N / 2)  # 片側スペクトル正規化 (N /2 で除算)

# sample frequencies (n: window length, d: sample spacing)
# f = [0, 1, ...,     n/2-1,     n/2] / (d*n)   if n is even
# f = [0, 1, ..., (n-1)/2-1, (n-1)/2] / (d*n)   if n is odd
xf_r = rfftfreq(n=N, d=d)

import plotly.graph_objects as go
fig = go.Figure()
fig.add_trace(go.Scatter(x=xf_r, y=yf_r_norm, name='周波数スペクトル (片側正規化)'))
fig.update_layout(showlegend=True, title='周波数スペクトル (scipy.rfft)', xaxis_title='周波数 [Hz]', yaxis_title='振幅')
fig.show()
```
