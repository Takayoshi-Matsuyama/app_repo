# 5. 周波数応答関数
$$
周波数応答関数 H(f) = \frac{出力波形の周波数スペクトル B(f)}{入力波形の周波数スペクトル A(f)} 
$$

※ $A(f)$、$B(f)$、$H(f)$ の値は複素数

引用：小野測器 FFT解析に関する 基礎用語集<br>
https://www.onosokki.co.jp/HP-WK/c_support/tech_term/cf_fft/cf3_3_2.htm

## 5.1 時系列データの要件

### 5.1.1 分析対象のスイープ周波数
1. スイープ開始周波数 $f_{start}$ = 0.1 [Hz]
2. スイープ終了周波数 $f_{end}$ = 10 [Hz]
3. 周波数分解能 $\Delta f_{req}$ = 0.1 [Hz]

### 5.1.2 時系列データの要件（最低限）
1. 周波数分解能 --> 最低限必要な時間（最小（最遅）周波数の波 1個分）
$$
T_{req-min} = \frac{1}{\Delta f_{req}} = \frac{1}{0.1} = 10 [\mathrm{s}]
$$

2. 最大分析周波数 --> 最低限必要なサンプリング周波数（最大（最速）周波数の2.56倍（※1））、時間刻み
$$
f_{req-min} = f_{end} \times 2.56 = 10 \times 2.56 = 25.6 [\mathrm{Hz}]  
$$
$$
\Delta t_{req-max} = \frac{1}{f_{req-min}} = \frac{1}{25.6} = 0.039625 [\mathrm{s}]
$$
3. 最低限必要なデータ点数
$$
N_{req-min} = T_{req-min} \times f_{req-min} = 10 \times 25.6 = 256 [個]
$$
4. 最低限必要なデータ点数（FFT要件 2のn乗）
$$
N_{fft-min} = N_{req}を2のn乗に切り上げ = 切り上げ(256) = 2^8 (そのまま)
$$

(※1) 2.56: エイリアシングの影響を低減するために、実際に使われる倍率<br>
引用：小野測器　周波数レンジとサンプリング周波数との関係は？<br>
https://www.onosokki.co.jp/HP-WK/c_support/faq/fft_common/fft_analys_1.htm

(※2) Note: 周波数 f [Hz]: 単位時間当たりのデータ個数 n [1/s]


### 5.1.3 時系列データの要件（分析精度向上）

分析精度を向上させるため、スイープ時間を長く（周波数分解能を高く）、サンプリング周波数を高く（高周波対応能力を高く）する。
1. スイープ時間 8倍（最低限必要な時間の3倍以上）
2. サンプリング周波数 10倍（最低限必要なサンプリング周波数の10倍以上）
3. データ点数 2のn乗

$$
T_{req} = T_{req-min} \times 4 = \frac{1}{\Delta f_{req}} \times 8 = \frac{1}{0.1} \times 8 = 80 [\mathrm{s}]
$$
$$
f_{req} = f_{end} \times 40.96 = 10 \times 40.96 = 409.6 [\mathrm{Hz}]  
$$
$$
\Delta t_{req} = \frac{1}{f_{req}} = \frac{1}{409.6} = 0.00244140625 [\mathrm{s}]
$$
$$
N_{req} = T_{req-min} \times f_{req} = 80 \times 409.6 = 32768 [個]
$$

### 5.2.6 入力と出力の周波数スペクトル

(OK) 推力の周波数スペクトルは、スイープ範囲（0.1 ～ 10Hz）で非ゼロ値。<br>
(OK) 物理オブジェクト位置の周波数スペクトルは、1Hz付近で最大となっている。（セクション「5.2.2 プラント（物理オブジェクト）の特性値を確認」の期待値と一致）

```python
#【FFT】FFT分析
from scipy.fft import rfft, rfftfreq  # 実数限定FFT
import numpy as np

# Number of sample points
N = len(mtnflow_df["time_s"][:-1])  # 最後の点を除き、32768点(2^15)にする
print(f"{N=}")

# sample spacing
d = flow.discrete_time.dt  # [s]
print(f"{d=}")

# spectrum (推力)
yf_force = rfft(mtnflow_df["force_N"][:-1])
yf_force_norm = np.abs(yf_force) / (N/2)  # 片側スペクトル正規化 (N/2で除算)
print(f"{yf_force.shape=}")

# spectrum (物理オブジェクト位置)
yf_obj_pos = rfft(mtnflow_df["obj_position_m"][:-1])
yf_obj_pos_norm = np.abs(yf_obj_pos) / (N/2)  # 片側スペクトル正規化 (N/2で除算)
print(f"{yf_obj_pos.shape=}")

# spectrum (物理オブジェクト位置)
yf_obj_vel = rfft(mtnflow_df["obj_velocity_m_s"][:-1])
yf_obj_vel_norm = np.abs(yf_obj_vel) / (N/2)  # 片側スペクトル正規化 (N/2で除算)
print(f"{yf_obj_vel.shape=}")

# sample frequencies (n: window length, d: sample spacing)
# f = [0, 1, ...,   n/2-1,     -n/2, ..., -1] / (d*n)   if n is even
# f = [0, 1, ..., (n-1)/2, -(n-1)/2, ..., -1] / (d*n)   if n is odd
xf = rfftfreq(n=N, d=d)

import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Scatter(x=xf, y=yf_force_norm, name='周波数スペクトル（推力）'))
fig.update_layout(showlegend=True, title='周波数スペクトル (推力)', xaxis_title='周波数 [Hz]', yaxis_title='振幅')
fig.update_xaxes(range=[0, 10])  # X軸をスイープ周波数範囲に制限
fig.show()

fig = go.Figure()
fig.add_trace(go.Scatter(x=xf, y=yf_obj_pos_norm, name='周波数スペクトル (物理オブジェクト位置)'))
fig.update_layout(showlegend=True, title='周波数スペクトル (物理オブジェクト位置)', xaxis_title='周波数 [Hz]', yaxis_title='振幅')
fig.update_xaxes(range=[0, 10])  # X軸をスイープ周波数範囲に制限
fig.show()

fig = go.Figure()
fig.add_trace(go.Scatter(x=xf, y=yf_obj_vel_norm, name='周波数スペクトル (物理オブジェクト速度)'))
fig.update_layout(showlegend=True, title='周波数スペクトル (物理オブジェクト速度)', xaxis_title='周波数 [Hz]', yaxis_title='振幅')
fig.update_xaxes(range=[0, 10])  # X軸をスイープ周波数範囲に制限
fig.show()
```

## 5.3 周波数応答関数

### 5.3.1 周波数応答関数を求める式（再掲）
$$
周波数応答関数 H(f) = \frac{出力波形の周波数スペクトル B(f)}{入力波形の周波数スペクトル A(f)} 
$$

※ $A(f)$、$B(f)$、$H(f)$ の値は複素数

### 5.3.2 チェック

* (OK) ゲインは1Hzで最大
* (OK) 正の周波数範囲において、位相は遅れていく

```python
# 位置の周波数応答関数
f_response_pos = yf_obj_pos / yf_force
f_response_pos_gain = np.abs(f_response_pos)
f_response_pos_phase = np.angle(f_response_pos)
f_response_pos_phase_unwapped = np.unwrap(f_response_pos_phase)
f_response_pos_phase_deg = np.degrees(f_response_pos_phase)
f_response_pos_phase_deg_unwrapped = np.degrees(f_response_pos_phase_unwapped)

# 速度の周波数応答関数
f_response_vel = yf_obj_vel / yf_force
f_response_vel_gain = np.abs(f_response_vel)
f_response_vel_phase = np.angle(f_response_vel)
f_response_vel_phase_unwrapped = np.unwrap(f_response_vel_phase)
f_response_vel_phase_deg = np.degrees(f_response_vel_phase)
f_response_vel_phase_deg_unwrapped = np.degrees(f_response_vel_phase_unwrapped)

from plotly.subplots import make_subplots
import plotly.graph_objects as go

# 位置 ゲイン・位相
fig = make_subplots(rows=2, cols=1, shared_xaxes=True, vertical_spacing=0.2, subplot_titles=('ゲイン', '位相'))
fig.add_trace(go.Scatter(x=xf, y=f_response_pos_gain, name='位置 周波数応答関数（ゲイン）'), row=1, col=1)
fig.add_trace(go.Scatter(x=xf, y=f_response_pos_phase_deg_unwrapped, name='位置 周波数応答関数（位相）'), row=2, col=1)
fig.update_yaxes(title='振幅', row=1, col=1)
fig.update_yaxes(title='位相 [deg]', row=2, col=1)
fig.update_xaxes(range=[0, 10], title_text='周波数 [Hz]')  # X軸をスイープ周波数範囲に制限
fig.update_layout(title='位置 周波数応答関数', showlegend=True, height=600)
fig.show()

# 速度 ゲイン・位相
fig = make_subplots(rows=2, cols=1, shared_xaxes=True, vertical_spacing=0.2, subplot_titles=('ゲイン', '位相'))
fig.add_trace(go.Scatter(x=xf, y=f_response_vel_gain, name='速度 周波数応答関数（ゲイン）'), row=1, col=1)
fig.add_trace(go.Scatter(x=xf, y=f_response_vel_phase_deg_unwrapped, name='速度 周波数応答関数（位相）'), row=2, col=1)
fig.update_yaxes(title='振幅', row=1, col=1)
fig.update_yaxes(title='位相 [deg]', row=2, col=1)
fig.update_xaxes(range=[0, 10], title_text='周波数 [Hz]')  # X軸をスイープ周波数範囲に制限
fig.update_layout(title='速度 周波数応答関数', showlegend=True, height=600)
fig.show()
```
