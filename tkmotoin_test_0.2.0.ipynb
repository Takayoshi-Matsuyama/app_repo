# インパルス関数プロファイルの確認

```python
import pprint
import tkmotion as tkm
flow = tkm.MotionFlow()
flow.load_discrete_time()
flow.load_plant()
print(f"{flow.module_version=}")
pprint.pprint(flow.discrete_time.get_config())
pprint.pprint(flow.plant.get_config())
```

```python
# motion_profileの確認 (default)
flow.load_motion_profile(filepath="tkmotion/prof/default_motion_prof.json", prof_index=1)
pprint.pprint(type(flow.mprof))
pprint.pprint(flow.mprof.get_config())
```

```python
# motion_profileの確認 (impulse)
flow.load_motion_profile(filepath="tkmotion/prof/default_motion_prof.json", prof_index=2)
pprint.pprint(type(flow.mprof))
pprint.pprint(flow.mprof.get_config())
```

```python
# controllerの確認 (default)
flow.load_controller(filepath="tkmotion/ctrl/default_controller.json", ctrl_index=1)
pprint.pprint(type(flow.controller))
pprint.pprint(flow.controller.get_config())
```

```python
# controllerの確認 (impulse)
flow.load_controller(filepath="tkmotion/ctrl/default_controller.json", ctrl_index=2)
pprint.pprint(type(flow.controller))
pprint.pprint(flow.controller.get_config())
```
