# Tech Note - NumPy complex

この技術ノートには、NumPyの複素数型の計算について記す。<br>
特に、周波数応答関数の「位相の符号」と「波の進み／遅れ」の関係を確認するために、複素数の除算を実行する。

```python
import numpy as np

# 複素数配列（指令値(1+0j)に対する、結果値を想定）
c_dummy_command_array = np.full(8, 1+0j)  # ダミー指令値配列（全て1+0jとする）
c_dummy_result_array = np.array([  # ダミー結果値配列
    1+0j,
    (1.1+1.1j)/np.sqrt(2),
    0+1.2j,
    (-1.3+1.3j)/np.sqrt(2),
    -1.4+0j,
    (-1.5-1.5j)/np.sqrt(2),
    0-1.6j,
    (1.7-1.7j)/np.sqrt(2)])

# プロット
import plotly.graph_objects as go
fig = go.Figure()
fig.add_trace(go.Scatter(
    x=c_dummy_result_array.real,
    y=c_dummy_result_array.imag,
    mode='markers'))
fig.update_layout(title='Dummy Results in the Complex Plane',
                  xaxis_title='Real Part',
                  yaxis_title='Imaginary Part',
                  xaxis=dict(scaleanchor="y", scaleratio=1),
                  width=600,
                  height=600)
fig.show()
```

```python
# 周波数応答の複素配列 = ダミー結果値配列 / ダミー指令値配列
# （指令値が基準なので、指令値が分母）
c_dummy_response_array = c_dummy_result_array / c_dummy_command_array
c_dummy_response_gain = np.abs(c_dummy_response_array)
c_dummy_response_phase = np.angle(c_dummy_response_array)
c_dummy_response_phase_deg = np.degrees(c_dummy_response_phase)
```

```python
# 周波数軸（ダミー）
index = np.arange(len(c_dummy_response_array))
index
```

```python
import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Scatter(x=index, y=c_dummy_response_gain, name='ダミー周波数応答関数（ゲイン）'))
fig.update_layout(showlegend=True, title='ダミー周波数応答関数（ゲイン）', xaxis_title='インデックス', yaxis_title='振幅')
fig.show()

fig = go.Figure()
fig.add_trace(go.Scatter(x=index, y=c_dummy_response_phase_deg, name='ダミー周波数応答関数（位相）'))
fig.update_layout(showlegend=True, title='ダミー周波数応答関数（位相）', xaxis_title='インデックス', yaxis_title='位相 [deg]')
fig.show()
```

位相がプラス（進み）：
* 位相が「プラス」= 基準よりも先んじて、進んで変化している。<br>
「0+1.2j」は、基準「1+0j」より「+90 deg」進んでいる。<br>
基準が0degのときに、"既に" 90degの状態。

位相がマイナス（遅れ）：
* 位相が「マイナス」= 基準よりも後に、遅れて変化している。<br>
「0-1.6j」は、基準「1+0j」より「-90 deg」遅れている。<br>
基準が0degのときに、"未だ" -90degの状態。
